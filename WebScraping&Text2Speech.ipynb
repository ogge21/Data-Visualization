### Web Scraping & Black Hole Text to Speech

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
page=requests.get('https://science.nasa.gov/astrophysics/focus-areas/black-holes')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
soup.prettify
```

```python
par=soup.find_all('p')[0]
```

```python
par.text
```

```python
from gtts import gTTS
text=par.text
output=gTTS(text=text,lang='en', slow=False)
output.save("blackhole.mp3")
```

```python
from IPython.display import Audio
audio=Audio(filename="blackhole.mp3")
audio
```
